```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ruta del archivo CSV (ajusta según el archivo que subas)
file_path = '/content/ia_generativa.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Crear el gráfico
plt.figure(figsize=(14, 8))  # Tamaño del gráfico

# Graficar cada columna de adopción por sector
for column in df.columns[2:]:
    plt.plot(df['Año'], df[column], marker='o', label=column)

# Graficar Tamaño del Mercado como una barra de fondo
plt.bar(df['Año'], df['Tamaño del Mercado (USD Billones)'], color='lightblue', alpha=0.4, label='Tamaño del Mercado (USD Billones)')

# Configurar los ticks del eje X para mostrar los años
plt.xticks(df['Año'], fontsize=12)
plt.xlabel("Año", fontsize=14)

# Configurar ticks del eje Y
plt.yticks(fontsize=12)
plt.ylabel("Porcentaje de Adopción y Tamaño del Mercado", fontsize=14)

# Títulos y etiquetas
plt.title("Evolución de la Inteligencia Artificial Generativa: Tamaño del Mercado y Usos", fontsize=18)

# Agregar leyenda
plt.legend(loc='upper left', fontsize=12)

# Cuadrícula para mayor claridad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño y mostrar
plt.tight_layout()
plt.show()
```
