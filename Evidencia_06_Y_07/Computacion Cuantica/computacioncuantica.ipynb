```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ruta del archivo CSV (ajusta según el archivo que subas)
file_path = '/content/computacion_cuantica.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Crear el gráfico
plt.figure(figsize=(14, 8))  # Tamaño del gráfico

# Graficar la inversión global
plt.plot(df['Año'], df['Inversión Global (USD Billones)'], marker='o', label='Inversión Global (USD Billones)', color='blue')

# Configurar los ticks del eje X para mostrar los años
plt.xticks(df['Año'], fontsize=12)
plt.xlabel("Año", fontsize=14)

# Configurar ticks del eje Y
plt.yticks(fontsize=12)
plt.ylabel("Inversión Global (USD Billones)", fontsize=14)

# Títulos y etiquetas
plt.title("Crecimiento de la Inversión Global en Computación Cuántica", fontsize=18)

# Agregar leyenda
plt.legend(loc='upper left', fontsize=12)

# Cuadrícula para mayor claridad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño y mostrar
plt.tight_layout()
plt.show()
```
