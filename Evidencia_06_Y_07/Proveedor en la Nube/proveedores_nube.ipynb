```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ruta del archivo CSV
file_path = '/content/proveedores_nube.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Crear el gráfico
plt.figure(figsize=(14, 8))  # Tamaño del gráfico

# Graficar cada proveedor de nube
for column in df.columns[1:]:  # Omitimos la columna de años
    plt.plot(df['AÃ±o'], df[column], marker='o', label=column)

# Configurar los ticks del eje X para mostrar los años
plt.xticks(df['AÃ±o'], fontsize=12)
plt.xlabel("Año", fontsize=14)

# Configurar ticks del eje Y
plt.yticks(fontsize=12)
plt.ylabel("Participación de Mercado (%)", fontsize=14)

# Títulos y etiquetas
plt.title("Evolución de Proveedores de Nube (2018-2022)", fontsize=18)

# Agregar leyenda
plt.legend(loc='upper left', fontsize=12)

# Cuadrícula para mayor claridad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño y mostrar
plt.tight_layout()
plt.show()
```
